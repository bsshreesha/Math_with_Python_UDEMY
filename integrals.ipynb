```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, latex
from IPython.display import display, Math

# Define the variable
x = symbols('x')

# Step 1: Define the function
f = x**2 + 2*x + 1
display(Math(f"f(x) = x^2 + 2x + 1"))

# Step 2: Compute the antiderivative (indefinite integral)
F = integrate(f, x)
display(Math(f"F(x) = \\int f(x) \\, dx = {latex(F)}"))

# Step 3: Evaluate the definite integral from 0 to 1
I = integrate(f, (x, 0, 1))
display(Math(f"I = \\int_0^1 f(x) \\, dx = {latex(I)}"))

# Step 4: Print result of definite integral
print(f"The value of the definite integral I is: {I}")

# Step 5: Visualize the function and shaded area under the curve
x_vals = np.linspace(-0.5, 1.5, 100)
y_vals = x_vals**2 + 2*x_vals + 1

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = $x^2 + 2x + 1$', color='blue')
plt.fill_between(x_vals, y_vals, where=(x_vals >= 0) & (x_vals <= 1), color='lightgray', alpha=0.5)
plt.xlim(-0.5, 1.5)
plt.ylim(0, 3)
plt.title('Area Under Curve for $f(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, latex, exp
from IPython.display import display, Math

# Define the variable
x = symbols('x')

# Step 1: Define the function for integration
f = x**2 * exp(5*x)
display(Math(r"I = \int x^2 e^{5x} \, dx"))

# Step 2: Apply Bernoulli's Rule for Integration by Parts
# Let u = x^2 and dv = e^(5x) dx
u = x**2
dv = exp(5*x)

# Compute derivatives and integrals
du = u.diff(x)  # u' = 2x
v = integrate(dv, x)  # v = (1/5)e^(5x)

# Display derivatives and integrals
display(Math(r"u = x^2, \, du = 2x \, dx"))
display(Math(r"dv = e^{5x} \, dx, \, v = \frac{1}{5} e^{5x}"))

# Step 3: Apply Bernoulli's formula
I = u * v - integrate(v * du, x)
display(Math(r"I = uv - \int v \, du"))
display(Math(r"I = x^2 \cdot \frac{1}{5} e^{5x} - \int \frac{1}{5} e^{5x} (2x) \, dx"))

# Step 4: Solve the remaining integral using integration by parts again
# Let u' = 2x and dv' = e^(5x)
u_prime = 2*x
dv_prime = exp(5*x)
du_prime = u_prime.diff(x)  # u'' = 2
v_prime = integrate(dv_prime, x)  # v' = (1/5)e^(5x)

# Display second derivatives and integrals
display(Math(r"u' = 2x, \, du' = 2 \, dx"))
display(Math(r"dv' = e^{5x} \, dx, \, v' = \frac{1}{5} e^{5x}"))

# Apply integration by parts again on the remaining integral
remaining_integral = integrate(v_prime * du_prime, x)
final_integral_result = I.subs(x, 'x') - remaining_integral

# Display final result of integral I
display(Math(r"I = \frac{1}{5} x^2 e^{5x} - \left(\frac{2}{25} e^{5x}(x - \frac{1}{5})\right) + C"))

# Final result after simplification (C is constant of integration)
final_result_simplified = final_integral_result.simplify()
display(Math(f"I = {latex(final_result_simplified)}"))

# Step 6: Visualize the function and shaded area under the curve for f(x)
x_vals = np.linspace(0, 0.5, 100)
y_vals = x_vals**2 * np.exp(5*x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = $x^2 e^{5x}$', color='blue')
plt.fill_between(x_vals, y_vals, color='lightgray', alpha=0.5)
plt.xlim(0, 0.5)
plt.ylim(0, np.max(y_vals) + 0.1)
plt.title('Area Under Curve for $f(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Function, Eq, dsolve, Derivative, exp, latex
from IPython.display import display, Math

# Step 1: Define the differential equation
x = symbols('x')  # Independent variable
y = Function('y')(x)  # Dependent variable
diffeq = Eq(Derivative(y, x), x + y)

# Print the differential equation
display(Math(r"\text{Differential Equation: } \frac{dy}{dx} = x + y"))

# Step 2: Solve the differential equation symbolically
solution = dsolve(diffeq, y)
display(Math(r"\text{Step 1: Solve the differential equation: } " + latex(solution)))
display(Math(r"\text{General Solution: } y(x) = C_1 e^x - x - 1"))

# Step 3: Apply initial condition y(0) = 1
y0 = 1  # Initial condition
particular_solution = dsolve(diffeq, y, ics={y.subs(x, 0): y0})
display(Math(r"\text{Step 2: Apply initial condition } y(0) = 1"))
display(Math(r"\text{Particular Solution: } " + latex(particular_solution)))

# Extract the particular solution
y_expr = particular_solution.args[1]

# Step 4: Step-by-step explanation for the solution
# (1) General solution: y(x) = C_1 e^x - x - 1
# (2) Apply initial condition y(0) = 1:
#     1 = C_1 e^0 - 0 - 1
#     1 = C_1 - 1
#     C_1 = 2
# (3) Substitute C_1 into the general solution:
#     y(x) = 2 e^x - x - 1
display(Math(r"""
\begin{aligned}
&\text{Step 3: Verify initial condition and find constant } C_1: \\
&y(0) = 1 \implies 1 = C_1 e^0 - 0 - 1 \\
&1 = C_1 - 1 \implies C_1 = 2 \\
&\text{Particular Solution: } y(x) = 2 e^x - x - 1
\end{aligned}
"""))

# Step 5: Numerical evaluation and visualization
x_vals = np.linspace(0, 3, 100)
y_vals = [(y_expr.subs(x, val).evalf()) for val in x_vals]

# Plot the solution
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=rf"$y(x) = {latex(y_expr)}$", color='blue')

# Enhance plot aesthetics
plt.title(r"Solution to $\frac{dy}{dx} = x + y$ with $y(0) = 1$")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
